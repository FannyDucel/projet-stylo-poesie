```python
from bs4 import BeautifulSoup as bs
import glob
import tqdm
import re
import json
```

```python
content = []
cpt = {}
dic_vers = {}
nb_docs = 0

for path in tqdm.tqdm(glob.glob("Mazarinades/*/*")):
    with open(path, "r", encoding="utf-8") as file:
        # Read each line in the file, readlines() returns a list of lines
        content = file.readlines()
        # Combine the lines in the list into a string
        content = "".join(content)
        bs_content = bs(content, "lxml")

    #il y a plusieurs dates : en priorité, on veut celle dans la balise <bibl>,
    #mais si on n'a pas de balise, on prend celle dans "setting"
    bibl = bs_content.find("bibl")
    setting = bs_content.find("setting")
    texte = bs_content.find("text")
    nb_docs += 1
    
    if bibl.find('date'):
        date = bibl.find("date")
    else:
        date = setting.find("date")
        
    if date.text != "Sans Date":
        date_doc = date.text.strip("([])")
        date_doc = date_doc[-4:]
    else:
        date_doc = "Sans Date"
        
    #garder les balises de type <term type="form">vers</term>, dates, texte
    #dico compteur vers/proses
    result = bs_content.find_all("term", attrs = {"type":"form"})
    for t in result:
        if date_doc not in cpt:
            cpt[date_doc] = {"vers":0,"prose":0}
        if t.text == "vers":
            cpt[date_doc]["vers"] = cpt[date_doc].get("vers",0)+1
            l = texte.find_all("l")
            for line in l:
                if path not in dic_vers:
                    dic_vers[path]=[]
                dic_vers[path].append(line.text.strip("\n"))
        else:
            cpt[date_doc]["prose"] = cpt[date_doc].get("prose",0)+1

print("Terminé !")
```

```python
print(cpt,"\n")

dic_proportions = {}
total_vers = 0
total_prose = 0

for date in cpt:
    dic_proportions[date] = {}
    tmp = cpt[date]["vers"] + cpt[date]["prose"]
    total_vers += cpt[date]["vers"]
    total_prose += cpt[date]["prose"]
    dic_proportions[date]["vers"] = cpt[date]["vers"]/tmp
    dic_proportions[date]["prose"] = cpt[date]["prose"]/tmp

print(dic_proportions,"\n")

proportion_totale_vers = total_vers/(total_vers+total_prose)
proportion_totale_prose = total_prose/(total_vers+total_prose)
print("Proportion de poèmes en vers:",proportion_totale_vers,"(%s textes sur %s)"%(total_vers,total_vers+total_prose))
print("Proportion de poèmes en prose:",proportion_totale_prose,"(%s textes sur %s)"%(total_prose,total_vers+total_prose))
```

```python
import matplotlib.pyplot as plt

prop_ordon = []
for date in cpt:
    prop_ordon.append([date,cpt[date]["vers"],cpt[date]["prose"]])  
prop_ordon = sorted(prop_ordon)
print(prop_ordon)

prop_ordon_rel = []
for date in dic_proportions:
    prop_ordon_rel.append([date,dic_proportions[date]["vers"],dic_proportions[date]["prose"]])  
prop_ordon_rel = sorted(prop_ordon_rel)
print(prop_ordon_rel)

points_x = []
points_y_vers_abs = []
points_y_vers_rel = []
points_y_prose_abs = []
points_y_prose_rel = []

for i in prop_ordon:
    points_x.append(i[0])
    points_y_vers_abs.append(i[1])
    points_y_prose_abs.append(i[2])

for i in prop_ordon_rel:
    points_y_vers_rel.append(i[1])
    points_y_prose_rel.append(i[2])
    
print("\n############\n\nPour les poèmes en vers (valeur absolue puis valeur relative):")
f = plt.figure()
f.set_figwidth(20)
plt.plot(points_x,points_y_vers_abs,label="vers")
plt.plot(points_x,points_y_prose_abs,label="prose")
plt.legend()
plt.show()

print("\n############\n\nPour les poèmes en valeur relative:")
f = plt.figure()
f.set_figwidth(20)
plt.plot(points_x,points_y_vers_rel,label="vers")
plt.plot(points_x,points_y_prose_rel,label="prose")
plt.legend()
plt.show()
```

```python

import json

def get_lexique_syll():
  with open("french-syllabification/syl_lexique.json") as f:
    dic_syll = json.load(f)
  return dic_syll
def indexes(word,letter):
    for i,x in enumerate(word):
             if x == letter:
                          yield i

def detect_syllables2(complete_text, dic_syll=get_lexique_syll()):

    word_text = []
    syls_list = []

    vowels = ['a', 'e', 'i', 'o', 'u', 'y','à', 'â', 'æ', 'œ', 'é', 'è', 'ê', 'ë', 'î', 'ï', 'ô', 'ù', 'û', 'ü', 'ÿ']
    punct = ['-', "'"]
    word_endings_2 = ['ez', 'is']
    word_endings_3 = ['ait', 'ons', 'ont']
    word_endings_4 = ['aient']
    word_endings_5 = ['issent', 'issons']

    complete_text = [str(x) for x in complete_text]
    known = set(complete_text).intersection(dic_syll.keys())
    for w in complete_text:
        syls = 0
        if w in known:
            syls_list.append(dic_syll[w])
        else:
            end_of_string_counter = 0
            two_cons_counter = 0
            last_ind = len(w)
            last_letter = w[(last_ind - 1)]
            for i,j, in enumerate(range(1,(len(w)))):
                if len(w) - j >= 2:
                    if w[i] in vowels:
                        if w[j] not in vowels and w[j] not in punct:
                            if w[j+1] in vowels:
                                test_result = list( indexes(w, w[j+1]) )
                                if test_result[-1] == (last_ind - 1):
                                   if end_of_string_counter == 0:
                                       if len(word_text) <= 5:
                                           syls += 1
                                           #print('end of string')
                                           end_of_string_counter = 1
                                       if len(word_text) >= 6:
                                           syls += 1
                                           #print('end of string')
                                if test_result[-1] != (last_ind - 1):
                                        if len(word_text) <= 6:
                                            syls += 1
                                            #print('one con added')
                                        if len(word_text) >= 7:
                                            syls += 1
                                            #print('one con added')

            # checking verb endings
            for x in word_endings_2:
                if w[-2:] == x:
                    check_con1 = w.split(x)
                    check_con1 = list(filter(None, check_con1))
                    check_con = str(check_con1)
                    if check_con[-1] not in vowels and check_con[-1] not in punct:
                        syls += 1
                    else:
                        syls += 1
            for x in word_endings_3:
                if w[-3:] == x:
                    check_con1 = w.split(x)
                    check_con1 = list(filter(None, check_con1))
                    check_con = str(check_con1)
                    if check_con[-1] not in vowels and check_con[-1] not in punct:
                        syls += 1
                    if check_con[-1] in vowels or check_con[-1] in punct:
                        syls += 1
            for x in word_endings_4:
                if w[-5:] == x:
                        check_con1 = w.split(x)
                        check_con1 = list(filter(None, check_con1))
                        check_con = str(check_con1)
                        if check_con[-1] not in vowels and check_con[-1] not in punct:
                            syls += 1
                        if check_con[-1] in vowels or check_con[-1] in punct:
                            syls += 1
                            #print('not added')
                            pass
            for x in word_endings_5:
                if w[-6:] == x:
                    w.split(x)
                    check_con = w.split(x)
                    if check_con[-1] not in vowels and check_con[-1] not in punct:
                        syls += 1
                    if check_con[-1] in vowels or check_con[-1] in punct:
                        pass

            # 2 vowels between two consonants
            for i,j, in enumerate(range(1,(len(w)))):
                    if len(w) - j >= 3:
                        if w[i] in vowels:
                            if w[j] not in vowels and w[j] not in punct:
                                if w[j+1] not in vowels and w[j+1] not in punct:
                                    if w[j+2] in vowels:
                                        if len(word_text) <= 6:
                                            syls += 1
                                            #print('two cons added')
                                            two_cons_counter = 1
                                        if len(word_text) > 6:
                                            syls += 1
                                            #print('two cons added')

            # 3 consonants between 2 vowels
            for i,j, in enumerate(range(1,(len(w)))):
                    if len(w) - j >= 4:
                        if w[i] in vowels:
                            if w[j] not in vowels and w[j] not in punct:
                                if w[j+1] not in vowels and w[j+1] not in punct:
                                    if w[j+2] not in vowels and w[j+2] not in punct:
                                        if w[j+3] in vowels and w[j+3] not in punct:
                                            if len(word_text) <= 6:
                                                syls += 1
                                                #print('three cons added')
                                            if len(word_text) > 6:
                                                syls += 1
                                                #print('three cons added')

            # consonants with two vowels in between
            for i,j, in enumerate(range(1,(len(w)))):
                    if len(w) - j >= 3:
                        if w[i] not in vowels:
                            if w[j] in vowels and w[j] not in punct:
                                if w[j+1] in vowels and w[j+1] not in punct:
                                        if w[j+2] not in vowels and w[j+2] not in punct:
                                                syls += 1

            # 3 vowels between 2 consonants
            for i,j, in enumerate(range(1,(len(w)))):
                    if len(w) - j >= 4:
                        if w[i] not in vowels:
                            if w[j] in vowels and w[j] not in punct:
                                if w[j+1] in vowels and w[j+1] not in punct:
                                    if w[j+2] in vowels and w[j+2] not in punct:
                                        if w[j+3] not in vowels and w[j+3] not in punct:
                                            if len(word_text) <= 6:
                                                syls += 1
                                                #print('three vowels added')
                                            if len(word_text) >= 7:
                                                syls += 1
                                                #print('three vowels added')

            #consonant followed by r or l
            for i,j, in enumerate(range(1,(len(w)))):
                if w[i] not in vowels:
                    if w[j] == 'r' or w[j] == 'l':
                        if w[j] != w[-1]:
                            if w[j+1] == 'e':
                                if w.index(w[j]) > 3:
                                    syls -= 1

            # dealing with hyphens, e aigu, and apostrophes
            for i,j, in enumerate(range(1,(len(w)))):
                if w[i] == '-':
                        syls += 1

            if w[-1:] == 'é':
                        syls += 1

            for i,j, in enumerate(range(1,(len(w)))):
                if w[i] == "'" and w[j] in vowels:
                        syls += 1

            # two or three vowels in a row (change to between consonants?)
            for i,j, in enumerate(range(1,(len(w)))):
                    if len(w) - j >= 4:
                        if w[i] in vowels:
                            if w[j] in vowels:
                                if w[j+1] in vowels:
                                    if w[j+2] not in vowels:
                                        pass
                                    else:
                                        if len(word_text) <= 6:
                                            syls += 1
                                            #print('two vowels subbed')
                                        if len(word_text) >= 7:
                                            syls += 1
                                            #print('two vowels subbed')
                                if w[j+1] not in vowels:
                                    if len(word_text) <= 6:
                                        pass
                                    if len(word_text) >= 7:
                                        syls += 1
                                        #print('two vowels subbed')

            if syls <= 0:
                if len(w) <= 5:
                    syls = 1
            syls_list.append(str(syls))

    rows = zip(complete_text, syls_list)
    liste_tuples = [row for row in rows]
    return liste_tuples

if __name__=="__main__":
  dic_syll = get_lexique_syll()
  res = []

  with open("dic_vers.json","r",encoding="utf-8") as f:
        docs = json.load(f)
  for chemin in tqdm.tqdm(docs):
    for vers in docs[chemin]:
      tmp = vers.split()
      res.append(detect_syllables2(tmp))
  
  toto = "Je suis toto un lapin présentement aérien avec beaucoup de qualités et certains menus défauts".split()
  print(res)
```
