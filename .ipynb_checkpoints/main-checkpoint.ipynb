### Fichier principal du projet : 
- proportion de poèmes en vers/prose avec figures et stats
- stockage des vers
- extraction et stockage des rimes

```python
from bs4 import BeautifulSoup as bs
import glob
import tqdm
import re
import json
```

```python
content = []
cpt = {}
dic_vers = {}
nb_docs = 0

for path in tqdm.tqdm(glob.glob("Mazarinades/*/*")):
    with open(path, "r", encoding="utf-8") as file:
        # Read each line in the file, readlines() returns a list of lines
        content = file.readlines()
        # Combine the lines in the list into a string
        content = "".join(content)
        bs_content = bs(content, "lxml")

    #il y a plusieurs dates : en priorité, on veut celle dans la balise <bibl>,
    #mais si on n'a pas de balise, on prend celle dans "setting"
    bibl = bs_content.find("bibl")
    setting = bs_content.find("setting")
    texte = bs_content.find("text")
    nb_docs += 1
    
    if bibl.find('date'):
        date = bibl.find("date")
    else:
        date = setting.find("date")
        
    if date.text != "Sans Date":
        date_doc = date.text.strip("([])")
        date_doc = date_doc[-4:]
    else:
        date_doc = "Sans Date"
        
    #garder les balises de type <term type="form">vers</term>, dates, texte
    #dico compteur vers/proses
    result = bs_content.find_all("term", attrs = {"type":"form"})
    for t in result:
        if date_doc not in cpt:
            cpt[date_doc] = {"vers":0,"prose":0}
        if t.text == "vers":
            cpt[date_doc]["vers"] = cpt[date_doc].get("vers",0)+1
            l = texte.find_all("l")
            for line in l:
                if path not in dic_vers:
                    dic_vers[path]=[]
                dic_vers[path].append(line.text.strip("\n"))
        else:
            cpt[date_doc]["prose"] = cpt[date_doc].get("prose",0)+1

print("Terminé !")
```

```python
print(cpt,"\n")

dic_proportions = {}
total_vers = 0
total_prose = 0

for date in cpt:
    dic_proportions[date] = {}
    tmp = cpt[date]["vers"] + cpt[date]["prose"]
    total_vers += cpt[date]["vers"]
    total_prose += cpt[date]["prose"]
    dic_proportions[date]["vers"] = cpt[date]["vers"]/tmp
    dic_proportions[date]["prose"] = cpt[date]["prose"]/tmp

print(dic_proportions,"\n")

proportion_totale_vers = total_vers/(total_vers+total_prose)
proportion_totale_prose = total_prose/(total_vers+total_prose)
print("Proportion de poèmes en vers:",proportion_totale_vers,"(%s textes sur %s)"%(total_vers,total_vers+total_prose))
print("Proportion de poèmes en prose:",proportion_totale_prose,"(%s textes sur %s)"%(total_prose,total_vers+total_prose))
```

```python
import matplotlib.pyplot as plt

prop_ordon = []
for date in cpt:
    prop_ordon.append([date,cpt[date]["vers"],cpt[date]["prose"]])  
prop_ordon = sorted(prop_ordon)
print(prop_ordon)

prop_ordon_rel = []
for date in dic_proportions:
    prop_ordon_rel.append([date,dic_proportions[date]["vers"],dic_proportions[date]["prose"]])  
prop_ordon_rel = sorted(prop_ordon_rel)
print(prop_ordon_rel)

points_x = []
points_y_vers_abs = []
points_y_vers_rel = []
points_y_prose_abs = []
points_y_prose_rel = []

for i in prop_ordon:
    points_x.append(i[0])
    points_y_vers_abs.append(i[1])
    points_y_prose_abs.append(i[2])

for i in prop_ordon_rel:
    points_y_vers_rel.append(i[1])
    points_y_prose_rel.append(i[2])
    
print("\n############\n\nPour les poèmes en vers (valeur absolue puis valeur relative):")
f = plt.figure()
f.set_figwidth(20)
plt.plot(points_x,points_y_vers_abs,label="vers")
plt.plot(points_x,points_y_prose_abs,label="prose")
plt.legend()
plt.show()

print("\n############\n\nPour les poèmes en valeur relative:")
f = plt.figure()
f.set_figwidth(20)
plt.plot(points_x,points_y_vers_rel,label="vers")
plt.plot(points_x,points_y_prose_rel,label="prose")
plt.legend()
plt.show()
```

```python
"""Total de rimes à l'échelle du corpus"""
import json
def trier_dic(dic):
    L = [[effectif,car] for car,effectif in dic.items()]
    L_sorted = sorted(L, reverse=True)
    return [[car,effectif] for effectif,car in L_sorted]

def total_rimes(path_dic_vers_rimes, path_json):
    with open(path_dic_vers_rimes, encoding="utf-8") as f:
        rimes = json.load(f)

    total_rimes = {}
    for chemin, dic_rimes in rimes.items():
        for rime, nb in dic_rimes.items():
            total_rimes[rime] = total_rimes.get(rime, 0) + nb
            
    with open(path_json,"w",encoding="utf-8") as f:
        json.dump(total_rimes,f)
        
    return total_rimes

total_rimes_2g = total_rimes("output/dic_vers_rimes_ngram2.json", "output/total_rimes_2g.json")
print("Top 20 des rimes en bigrammes les plus fréquentes", trier_dic(total_rimes_2g)[:20])

total_rimes_3g = total_rimes("output/dic_vers_rimes_ngram3.json", "output/total_rimes_3g.json")
print("\nTop 20 des rimes en trigrammes les plus fréquentes",trier_dic(total_rimes_3g)[:20])
```

```python
"""Modification pour récupérer la liste de rimes par fichier"""
import glob, string

def get_rimes(vers,ngram):  #prend en entrée une liste de vers et le nombre caractères de fin de vers à comparer
    rimes = []
    
    for ligne in vers:
        ligne_clean = ligne.translate(str.maketrans('','',string.punctuation))
        rimes.append(ligne_clean[-ngram:])
    
    return rimes

with open("output/dic_vers_corrige.json", encoding="utf-8") as f:
    dic_rimes = json.load(f)

dic_rimes_detail = {}

for path, docs in dic_rimes.items():
    rimes = get_rimes(docs,2)
    dic_rimes_detail[path]=[rime.lower() for rime in rimes]

with open("output/dic_vers_rimes_2g_detail.json","w",encoding="utf-8") as f:
    json.dump(dic_rimes_detail,f)
```

```python
"""Établir le type de rimes (suivies, croisées, embrassées)
: pour bigrammes"""

rimes_s, rimes_c, rimes_e, no_rimes, total = 0,0,0,0, 0
for path, liste_rimes in dic_rimes_detail.items():
    for i, rime in enumerate(liste_rimes):
        total += 1
        if i<len(liste_rimes)-1 and rime == liste_rimes[i+1]:
            type_rime = "s" #si la rime est la même que la suivante (AA), c'est une rime suivie
            rimes_s +=1
        elif i<len(liste_rimes)-2 and rime == liste_rimes[i+2]:
            type_rime = "c" #si la rime est la même que 2 lignes en-dessous (ABA), c'est une rime croisée
            rimes_c +=1
        elif i<len(liste_rimes)-3 and rime == liste_rimes[i+3]:
            type_rime = "e" #si la rime est la même que 3 lignes en-dessous (ABBA), c'est une rime embrassée 
            rimes_e +=1
        else:
            type_rime = "n" #n pour non, pas de rime
            no_rimes +=1

print("Il y a ", rimes_s, "rimes suivies (soit ", round(rimes_s/total*100,2), "%), ", rimes_c, 
      "rimes croisées (soit ", round(rimes_c/total*100,2), "%),", rimes_e, "rimes embrassées (soit",
      round(rimes_e/total*100,2), "%) et", no_rimes, "vers sans rimes (soit", round(no_rimes/total*100,2) ,"%).")

"""Limite : une rime embrassée sera donc comptée comme embrassée pour A et suivies pour B
+ 1 rime = pour deux vers donc si un vers est écrit uniquement en rimes suivies, il y aura autant de rimes que de vers
+ problème car dans nos filtres nous ignorons certains vers (bruités)"""
```

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ["Rimes suivies", "Rimes croisées", "Rimes embrassées","Vers sans rimes"]
values = [rimes_s, rimes_c, rimes_e, no_rimes]
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(labels, values, width = 0.4)
 
plt.xlabel("Type de rimes")
plt.ylabel("Nombre de rimes")
plt.show()
```
