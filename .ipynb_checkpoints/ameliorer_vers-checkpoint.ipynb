```python
"""Modifier les vers et créer un nouveau dic_vers.json avec quelques modifications/corrections 
pour améliorer découpe syllabique ensuite : 
- [OK] s longs
- [OK] dissimilation (v > u si le caractère d'après est une consonne ou espace [= dernier car du mot, pas de mot 
finissant en v])
- marque de césure (réutiliser code TD)
- [OK] ENLEVER vers trop courts/avec chiffres
- [OK] enlever vers avec répétition de lettres à la suite ? (= exclure problèmes d'OCR/bruit)
"""
import json

def has_numbers(inputString): #https://stackoverflow.com/questions/19859282/check-if-a-string-contains-a-number
    return any(char.isdigit() for char in inputString)

def quadruple_letters(word): 
    """adapté de https://stackoverflow.com/questions/69888959/checking-string-for-double-letter-in-a-row
    pour détecter mots ayant + de 4x le même caractère à la suite (sauf si ce caractère est ponctuation, espace, césure)
    on fait avec 4 lettres car possible et commun d'avoir 2 fois la même lettre à la suite (ss, ll, ée)
    et parfois 3 lettres pareilles à la suite, erreur mais pas très grave, comprend quand même, pas d'impact sur syllabation
    /!/ à faire à la fin
    """
    for i in range (len(word)-3):
        if word[i] == word[i+1] == word[i+2] == word[i+3] and word[i].isalpha():
            return True, word[i]
    return False

import re 
def dissimilation(vers):
    """On remplace les v par des u si la lettre suivant le v est une consonne ou si v est le dernier car du mot"""
    if "v" in vers:
        indices = [i.start() for i in re.finditer("v", vers)]
        for j in indices:
            if j == len(vers)-1:
                #peut pas utiliser replace car remplace tous les caractères indiqués, pas selon un index
                vers = vers[:j] + "u" + vers[j + 1:]
            elif vers[j+1] not in "aeiouy":
                vers = vers[:j] + "u" + vers[j + 1:]
    return vers

def gerer_cesure(liste_vers):
    """césure en fin de ligne marquée par ¬, on gère seulement cas où césure en fin de vers,
    alors vers continue en fait sur ligne suivante donc on recolle ensemble le vers suivant"""
    nv_lv = []
    j = 0
    for i, vers in enumerate(liste_vers):
        if len(vers)>0 and vers[-1]=="¬":
            try:
                v = vers[:-1]+liste_vers[i+1]
            except IndexError: #problème quand césure en fin de page
                pass
            j = i+1
        else:
            v = vers
        if j != i or j==0: #pour ne pas remettre les vers "suites" (ceux qui ont été recollés) + j==0 sinon tous les 1ers vers sautent
            nv_lv.append(v)
    return nv_lv


with open("output/dic_vers.json", encoding="utf-8") as f:
    dic_vers = json.load(f)

nv_dic_vers = {}

for chemin, liste_vers in dic_vers.items():
    lv = gerer_cesure(liste_vers)
    #garder vers seulement s'il contient + de 5 car et pas tout en maj (garde chiffre finalement pcq si un seul dans une vraie chaîne, reste compréhensible)
    lv = [vers for vers in lv if len(vers)>5 and vers.isupper() == False] #and has_numbers(vers)==False
    #remplacer s longs
    lv = [vers.replace("ſ","s") for vers in lv]
    #remplacer v par u si lettre suivante est une consonne ou s'il s'agit du dernier caractère
    lv = [dissimilation(vers) for vers in lv]
    #enlever vers qui contiennent + de 4 lettres répétées à la suite
    lv = [vers for vers in lv if quadruple_letters(vers)==False]
    #enlever espaces de début et fin
    lv = [vers.strip() for vers in lv]
    nv_dic_vers[chemin]=lv
    
with open('output/dic_vers_corrige.json', 'w') as f:
    json.dump(nv_dic_vers, f, indent=4)
```
